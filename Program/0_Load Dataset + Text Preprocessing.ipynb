# Text Preprocessing

```python
import nltk
import re
import numpy as np
import spacy

stop_words = nltk.corpus.stopwords.words('english')
# just to keep negation if any in bi-grams
stop_words.remove('no')
stop_words.remove('not')

nlp = spacy.load('en', disable=['vectors', 'textcat', 'tagger', 'parser', 'ner'])

def normalize_document(doc):
    # lower case and remove special characters\whitespaces
    doc = re.sub(r'[^a-zA-Z\s]', " ", doc)
    doc = doc.lower()
    doc = doc.strip()
    # tokenize document
    tokens = nltk.word_tokenize(doc)
    # filter stopwords out of document
    filtered_tokens = [token for token in tokens if token not in stop_words]
    # lemmatize the words
    lemma_tokens = []
    for token in filtered_tokens:
        text  = nlp(token)
        for one_word in text:
            if one_word.lemma_ != '-PRON-':
                lemma_tokens.append(one_word.lemma_)
            else:
                lemma_tokens.append(one_word.text)
    # re-create document from lemmatized tokens
    doc = ' '.join(lemma_tokens)
    return doc

#normalize_corpus = np.vectorize(normalize_document)
#norm_corpus = normalize_document(list(PubMed['Description']))

desc_list = PubMed['Description'].to_list()
norm_corpus = [normalize_document(item) for item in desc_list]

print(norm_corpus[:2])
```

```python
PubMed['Clean_Description'] = norm_corpus
PubMed
```

```python
pd.DataFrame(PubMed, columns = ['PUMID', 'Title', 'Abstract', 'CancerType', 'Description', 'Clean_Description']).to_excel('../Src/(Cleaned)pubmed-CancerType_Top1-10-set_10000-data.xlsx', index=False)
```
